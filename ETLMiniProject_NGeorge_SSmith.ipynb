## **1.2 Extract Categories and SubCategories From One Column**

### Crowdfunding DataFrame's Column Title List

```python
log_subroutine \
    .PrintAndLogWriteText \
        ('\033[1m' \
         + "This is a List of the Crowdfunding DataFrame's column titles:\n\n" \
         + '\033[0m' \
         + str \
             (crowdfundingDataFrame.keys()))
```

### Category and Subcategory Separation

```python
# These lines of code separate the 'category & sub-category' column into 'category' 
# and 'subcategory' columns before droppping the original if the 'category & 
# sub-category' column exits, of course.

if 'category & sub-category' in crowdfundingDataFrame.columns:

    crowdfundingDataFrame \
        [['category', 'subcategory']] \
            = crowdfundingDataFrame \
                ['category & sub-category'] \
                    .str \
                    .split \
                        ('/', 
                         expand = True)

    crowdfundingDataFrame \
        .drop \
            ('category & sub-category', 
             axis = 1, 
             inplace = True)

    log_function \
        .DebugReturnObjectWriteObject \
            (crowdfundingDataFrame)
```

### Display Updated Crowdfunding DataFrame

```python
captionString \
    = 'Table 1.2.1: Updated Crowdfunding DataFrame'

currentStylerObject \
    = function \
        .ReturnStylerObjectStandardFormat \
            (crowdfundingDataFrame.head(12),
             captionString,
             hideFlagBooleanParameter = True) \
                .format \
                    ({'cf_id': '{:}',
                      'contact_id': '{:}',
                      'goal': '{:,.2f}',
                      'pledged': '{:,.2f}',
                      'backers_count': '{:,}'})

log_function \
    .ReturnStylerObjectSavePNGImage \
        (currentStylerObject,
         captionString)
```

### Category List

```python
# This line of code extracts the unique categories, sorts them alphabetically, 
# and places them in a List.
categoriesStringList \
    = sorted \
        (crowdfundingDataFrame \
            ['category'] \
                 .unique() \
                 .tolist())

log_subroutine \
    .PrintAndLogWriteText \
        ('\033[1m' \
         + 'This is the list of {:,} categories from the Crowdfunding DataFrame ' \
                .format \
                 (len \
                      (categoriesStringList)) \
         + 'in alphabetical order:\n\n' \
         + '\033[0m' \
         + str \
             (categoriesStringList))
```

### Sub-Category List

```python
# This line of code extracts the unique categories, sorts them alphabetically, 
# and places them in a List.
subCategoriesStringList \
    = sorted \
        (crowdfundingDataFrame \
             ['subcategory'] \
                 .unique() \
                 .tolist())

log_subroutine \
    .PrintAndLogWriteText \
        ('\033[1m' \
         + 'This is the list of {:,} subcategories from the Crowdfunding DataFrame '
                .format \
                    (len \
                        (subCategoriesStringList)) \
         + 'in alphabetical order:\n\n' \
         + '\033[0m' \
         + str \
             (subCategoriesStringList))
```

### Category and Subcategory IDs

```python
# These lines of code create and display numPy arrays from 1-9 for category IDs
# and 1-24 for subcategory IDs.
categoryArrayLimitInteger \
    = len \
        (categoriesStringList) \
      + 1

subCategoryArrayLimitInteger \
    = len \
        (subCategoriesStringList) \
      + 1


categoryIDsNumPyArray \
    = np \
        .arange \
            (1, categoryArrayLimitInteger)

subCategoryIDsNumPyArray \
    = np \
        .arange \
            (1, subCategoryArrayLimitInteger)


log_subroutine \
    .PrintAndLogWriteText \
        ('\033[1m' \
         + 'This is the numPy Array for Category IDs:\n'
         + '\033[0m' \
         + str \
             (categoryIDsNumPyArray) \
         + '\n')

log_subroutine \
    .PrintAndLogWriteText \
        ('\033[1m' \
         + 'This is the numPy Array for Subcategory IDs:\n'
         + '\033[0m' \
         + str \
             (subCategoryIDsNumPyArray))
```

## **1.3 Create and Transform Category and SubCategory DataFrames**

### Create Category DataFrame

```python
crowdfundingCategoryDataFrame \
    = pd \
        .DataFrame \
            ({'category_id': 
                  categoryIDsNumPyArray,
              'category': 
                  categoriesStringList})
```

### Update Category ID with Prefix

```python
# These lines of code use a list comprehension to add 'cat' to each 'category_id' value. 
crowdfundingCategoryDataFrame \
    ['category_id'] \
        = 'cat' \
          + crowdfundingCategoryDataFrame \
                ['category_id'] \
                    .astype \
                        (str)


log_function \
    .DebugReturnObjectWriteObject \
        (crowdfundingCategoryDataFrame)
```

### Display Category DataFrame

```python
captionString \
    = 'Table 1.3.1: Category DataFrame'

currentStylerObject \
    = function \
        .ReturnStylerObjectStandardFormat \
            (crowdfundingCategoryDataFrame,
             captionString,
             hideFlagBooleanParameter = True)

log_function \
    .ReturnStylerObjectSavePNGImage \
        (currentStylerObject,
         captionString)
```

### Create Subcategory DataFrame

```python
crowdfundingSubCategoryDataFrame \
    = pd.DataFrame \
        ({'subcategory_id': subCategoryIDsNumPyArray,
          'subcategory': subCategoriesStringList})
```

### Update Subcategory ID with Prefix

```python
# These lines of code use a list comprehension to add 'cat' 
# to each 'category_id' value. 
crowdfundingSubCategoryDataFrame \
    ['subcategory_id'] \
        = 'subcat' \
          + crowdfundingSubCategoryDataFrame \
                ['subcategory_id'] \
                    .astype(str)


log_function \
    .DebugReturnObjectWriteObject \
        (crowdfundingSubCategoryDataFrame)
```

### Display Subcategory DataFrame

```python
captionString \
    = 'Table 1.3.2: Subcategory DataFrame'

currentStylerObject \
    = function \
        .ReturnStylerObjectStandardFormat \
            (crowdfundingSubCategoryDataFrame,
             captionString,
             hideFlagBooleanParameter = True)

log_function \
    .ReturnStylerObjectSavePNGImage \
        (currentStylerObject,
         captionString)
```

## **1.4 Export Category and Subcategory DataFrames to CSV Files**

```python
crowdfundingCategoryDataFrame \
    .to_csv \
        (local_constant \
             .CONSTANT_CATEGORY_FILE_PATH, 
         index = False)

crowdfundingSubCategoryDataFrame \
    .to_csv \
        (local_constant \
             .CONSTANT_SUBCATEGORY_FILE_PATH, 
         index = False)
```

### Clean DataFrame by Dropping Unwanted Columns

# <br> **3. Contacts DataFrame**

## **3.1 Create Contacts DataFrame**

### Initial Contacts DataFrame from Excel Workbook

```python
# This function reads the data from an Excel Workbook and assigns it 
# to a DataFrame.
contactsDataFrame \
    = function \
        .ReturnExcelFileAsDataFrame \
            (local_constant \
                 .CONSTANT_CONTACTS_EXCEL_FILE_PATH,
             'contact_info',
             headerRowInteger \
                = 3,
             dataTypeDictionary \
                = local_constant \
                     .excelContactsDTypeDictionary)


log_function \
    .DebugReturnObjectWriteObject \
        (contactsDataFrame)
```

## **3.2 Extract Columns With Regular Expressions**

### Extract Contact ID And Add New Column

```python
searchString \
    = '(\d{4})'

contactsDataFrame \
    ['contact_id'] \
        = contactsDataFrame \
            ['contact_info'] \
                .str \
                .extract \
                    (searchString)


log_function \
    .DebugReturnObjectWriteObject \
        (searchString \
         + '\n\n' 
         + str \
             (contactsDataFrame \
                ['contact_info']))
```

### Display Updated Contacts DataFrame With Contact ID

```python
captionString \
    = 'Table 3.2.1: Updated Contacts DataFrame with Contact ID'

currentStylerObject \
    = function \
        .ReturnStylerObjectStandardFormat \
            (contactsDataFrame.head(12),
             captionString,
             hideFlagBooleanParameter = True) \
                .format \
                    ({'contact_id': '{:}'})

log_function \
    .ReturnStylerObjectSavePNGImage \
        (currentStylerObject,
         captionString)
```

### Contacts DataFrame Information

```python
log_subroutine \
    .PrintAndLogWriteText \
        ('\033[1m' \
         + 'Table 3.2.2: Contacts DataFrame Information:\n' 
         + '\033[0m')

contactsDataFrame \
    .info \
        (verbose = True,
         memory_usage = True, 
         show_counts = True)
```

### Convert Appropriate Column to `int64` Data Type

```python
contactsDataFrame \
    ['contact_id'] \
        = contactsDataFrame['contact_id'] \
                .astype(np.int64)
    
    
log_function \
    .DebugReturnObjectWriteObject \
        (str \
             (type \
                 (contactsDataFrame \
                    ['contact_id'])) \
         + '\n\n' \
         + str \
             (contactsDataFrame \
                ['contact_id']))
```

### Contacts DataFrame Information with Updated Column Data Type

```python
log_subroutine \
    .PrintAndLogWriteText \
        ('\033[1m' \
         + 'Table 3.2.3: Contacts DataFrame Information ' \
         + 'with Updated Column Data Type:\n' 
         + '\033[0m')

contactsDataFrame \
    .info \
        (verbose = True,
         memory_usage = True, 
         show_counts = True)
```

### Extract Name And Add New Column

```python
searchString \
    = r'([^name"\s][a-zA-Z]+\s+[a-zA-Z]+)'

contactsDataFrame['name'] \
        = contactsDataFrame['contact_info'] \
            .str \
            .extract \
                (searchString)


log_function \
    .DebugReturnObjectWriteObject \
        (searchString \
         + '\n\n' 
         + str \
             (contactsDataFrame \
                ['contact_info']))
```

### Display Updated Contacts DataFrame With Name

```python
captionString \
    = 'Table 3.2.4: Updated Contacts DataFrame With Name'

currentStylerObject \
    = function \
        .ReturnStylerObjectStandardFormat \
            (contactsDataFrame.head(12),
             captionString,
             hideFlagBooleanParameter = True) \
                .format \
                    ({'contact_id': '{:}'})

log_function \
    .ReturnStylerObjectSavePNGImage \
        (currentStylerObject,
         captionString)
```

### Extract E-mail And Add New Column

```python
searchString \
    = r'"(\S+@\S+)"}'

contactsDataFrame['email'] \
        = contactsDataFrame['contact_info'] \
            .str \
            .extract \
                (searchString)


log_function \
    .DebugReturnObjectWriteObject \
        (searchString \
         + '\n\n' 
         + str \
             (contactsDataFrame \
                ['contact_info']))
```

### Display Updated Contacts DataFrame With E-mail

```python
captionString \
    = 'Table 3.2.5: Updated Contacts DataFrame With E-mail'

currentStylerObject \
    = function \
        .ReturnStylerObjectStandardFormat \
            (contactsDataFrame.head(12),
             captionString,
             hideFlagBooleanParameter = True) \
                .format \
                    ({'contact_id': '{:}'})

log_function \
    .ReturnStylerObjectSavePNGImage \
        (currentStylerObject,
         captionString)
```
